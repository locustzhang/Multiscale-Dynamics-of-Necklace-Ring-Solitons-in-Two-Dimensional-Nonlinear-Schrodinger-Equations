```python
import json
import os

# ==========================
# 读取函数
# ==========================
def load_json(fname):
    if not os.path.exists(fname):
        print(f"⚠️  未找到：{fname}")
        return None
    with open(fname, 'r', encoding='utf-8') as f:
        return json.load(f)

# ==========================
# 加载你真实的 4 个文件
# ==========================
scan1d = load_json("scan1d.json")
convergence = load_json("convergence.json")
scan2d = load_json("scan2d.json")
collision_scan = load_json("collision_scan.json")

# ==========================
# 开始输出真实数据
# ==========================

print("="*80)
print("📌 1D 吸引相互作用扫描 (scan1d.json)")
print("="*80)
if scan1d:
    for item in scan1d:
        print(f"g = {item['g']:.3f} | ρ_ratio = {item['rho_ratio']:.3f} | N/Ncr = {item['N_over_Ncr']:.3f} | E_err = {item['E_err']:.2e}")

print("\n" + "="*80)
print("📌 数值收敛测试 (convergence.json)")
print("="*80)
if convergence:
    print("\n→ 时间收敛 (dt sweep):")
    for r in convergence["dt_results"]:
        print(f"dt={r['dt']:.0e} | sol_err={r['sol_err']:.2e} | E_err={r['E_err']:.2e}")

    print("\n→ 空间收敛 (dx sweep):")
    for r in convergence["nx_results"]:
        print(f"Nx={r['Nx']} | dx={r['dx']:.3f} | E_err={r['E_err']:.2e}")

print("\n" + "="*80)
print("📌 2D 稳定性相图 (scan2d.json)")
print("="*80)
if scan2d:
    total = 0
    collapsed = 0
    for row in scan2d:
        for p in row:
            total +=1
            if p['collapsed']:
                collapsed +=1
    print(f"总参数点：{total} | 塌缩点数：{collapsed} | 塌缩比例：{collapsed/total*100:.1f}%")

print("\n" + "="*80)
print("📌 双环碰撞分类 (collision_scan.json)")
print("="*80)
if collision_scan:
    for res in collision_scan:
        print(f"v={res['v']:.1f} | 类型={res['type']:15s} | ρ_ratio={res['rho_ratio']:6.3f} | 峰数={res['n_peaks']:2d} | 径向相关={res['radial_corr']:.3f}")

print("\n" + "="*80)
print("✅ 4 个 JSON 全部提取完成！")
print("="*80)
```
